```python
import matplotlib.pyplot as plt

x = [1,2,3,4,7]
y = [0,4,2,3,6]
y2 = [2,4,5,4,5]

plt.plot(x,y,color='blue',label='Training Accuracy',linestyle='dashed',marker='*')
plt.plot(x,y2,color='red',label='Validation Accuracy',marker='o')
plt.title("Accuracy plot")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")

plt.legend(title="Training the Model")
plt.show()

```

```python
import numpy as np

x = np.random.rand(50)
y = np.random.rand(50)

y2 = np.random.rand(50)*10


plt.scatter(x,y,color='blue',label='blue_points',marker='*',alpha=0.9)
plt.scatter(x,y2,color='red',label='red_points',marker='o')
plt.legend()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter Plot")
plt.show()

```

```python
x = ['A', 'B', 'C']
y = [10, 5, 7]

plt.bar(x,y,label=['A','B','C'],color=['red','green','blue'])
plt.legend()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Bar chart")
plt.show()
```

```python
x = np.linspace(0,20,100)
y = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y)
plt.plot(x,y2)

plt.show()
```
