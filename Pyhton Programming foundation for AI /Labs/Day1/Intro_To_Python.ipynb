# 4. Data Structure:

## 4.1 Lists

Lists are a container type for ordered sequences of elements. Lists can be initialized empty.
elements-in-python-list-1024x495.jpg

```python
my_list = []
```

or with some initial elements

```python
my_list = [1, 2, 3]
```

can be from the same data type or different data type

```python
my_list = [1, 2, 3] # Integers
my_list = [1, 2, 'three'] # Integers and string
```

Lists have many properties and functions. e.x


```python
list_var = [1, 2, 3, 4]
print("List:", list_var)
```

```python
# Lists have a dynamic size
print("Length:", len(list_var))
```

```python
# We can access individual elements of a list (indexing starts from 0)
print("Access first element:", list_var[0])
```

```python
# We can access "slices" of a list using `my_list[i:j]` where `i` is the start of the slice (again, indexing starts from 0) and `j` the end of the slice. For instance:
print("Slice [1:3]:", list_var[1:3])
```

```python
# Omitting the second index means that the slice shoud run until the end of the list
print("Slice [1:]:", list_var[1:])
```

```python
# We can check if an element is in the list using `in`
print(5 in list_var)
```

```python
# elements can be added (appended) to them
list_var.append(5)
print("After appending 5:", list_var)
```

```python
# can insert element in specific index
list_var.insert(1, 1.5)
print("After inserting 1.5 at index 1:", list_var)
```

```python
# or remove elemnt from a specific index
list_var.remove(2)
print("After removing 2:", list_var)
```

```python
# or pop the lst element
popped_element = list_var.pop(0)
print("Popped element:", popped_element)
print("List after popping:", list_var)
```

```python
# list can be sorted ascending or descending
list_var.sort()
print("Sorted list ascending:", list_var)
list_var.sort(reverse=True)
print("Sorted list descending:", list_var)

list_var.reverse()
print("Reversed list:", list_var)
```

## 4.2 Tuples
Definition: A tuple is an ordered, immutable collection of elements, defined with parentheses ().

```python
# Tuples are like lists, but they are immutable
tuple_var = (1, 2, 3, 'four')

print("Tuple:", tuple_var)
print("Length:", len(tuple_var))
print("Access first element:", tuple_var[0])
print("Slice [1:3]:", tuple_var[1:3])
print("Count of 'four':", tuple_var.count('four'))
print("Index of 2:", tuple_var.index(2))
```

## 4.3 Set
Definition: A set is an unordered collection of unique elements, defined with curly braces {}, it has no Indexing or Slicing.

```python
# 1. Unordered
s = {1, 2, 3}
print(s)  # Output: {1, 2, 3} (order may vary)

# 2. Unique Elements
s_unique = {1, 2, 2, 3}
print(s_unique)  # Output: {1, 2, 3} (duplicates are removed)

# 3. Mutable
s_mutable = {1, 2, 3}
s_mutable.add(4)  # You can add elements
print(s_mutable)  # Output: {1, 2, 3, 4}

# 4. Can Store Multiple Data Types
s_mixed = {1, "hello", 3.5}
print(s_mixed)  # Output: {1, 'hello', 3.5}
```

```python
# Dictionaries store data in key-value pairs
dict_var = {'key1': 'value1', 'key2': 'value2'}
print("Dictionary:", dict_var)
print("Length:", len(dict_var))
# Accessing dictionary elements
print("Access value for 'key1':", dict_var['key1'])
# Get dictionary keys
print("Keys:", dict_var.keys())
# Get dictionary values
print("Values:", dict_var.values())
# Get dictionary items
print("Items:", dict_var.items())
print("Get value for 'key1':", dict_var.get('key1'))
removed_value = dict_var.pop('key1')
print("Removed value for 'key1':", removed_value)
print("Dictionary after removing 'key1':", dict_var)
dict_var.update({'key3': 'value3'})
print("Dictionary after adding key3:", dict_var)
dict_var.clear()
print("Dictionary after clearing:", dict_var)
```

# 5. Operations:
Definition: Actions or processes that manipulate data or perform calculations.



## 5.1 Arithmetic operations

Python supports the usual arithmetic operators: + (addition), * (multiplication), / (division), ** (power), // (integer division).

```python
a = 15
b = 4

print(a + b)  # Addition
print(a - b)  # Subtraction
print(a * b)  # Multiplication
print(a / b)  # Division (float result)
print(a // b)  # Floor division
print(a % b)  # Modulus
print(a ** b)  # Exponentiation
```

## 5.2 Assignment Operators

```python
# Assignment operators are used to assign values to variables
a = 10  # Basic assignment
print(a)
a += 5  # Equivalent to a = a + 5
print(a)
a -= 3  # Equivalent to a = a - 3
print(a)
a *= 2  # Equivalent to a = a * 2
print(a)
a /= 2  # Equivalent to a = a / 2
print(a)
```

## 5.3 Comparison Operators

```python
# Comparison operators compare two values and return True or False
x = 10
y = 5

print(x == y)  # Equal
print(x != y)  # Not equal
print(x > y)  # Greater than
print(x < y)  # Less than
print(x >= y)  # Greater than or equal to
print(x <= y)  # Less than or equal to
```

## 5.4 Logical Operators


```python
# Logical operators combine conditional statements
x = True
y = False

print(x and y)  # True if both are True
print(x or y)  # True if at least one is True
print(not x)  # True if x is False

print(30*'-')
print(x & y)  # True if both are True
print(x | y)  # True if at least one is True
```

##  5.5 Membership Operators

```python
# Membership operators check if a sequence is present in an object
fruits = ["apple", "banana", "cherry"]

print("banana" in fruits)  # Returns True because "banana" is in the list
print("orange" not in fruits)  # Returns True because "orange" is not in the list
```

# 6. Conditional Statements(If Statement):
* Definition: An if statement is used to execute a block of code only if a specified condition is true. It helps make decisions in a program.



if-statement.png

* Example:

As their name indicates, conditionals are a way to execute code depending on whether a condition is True or False. As in other languages, Python supports `if` and `else` but `else if` is contracted into `elif`, as the example below demonstrates.

```python
if x > 0:
    print("x is positive")  # This code runs only if x is greater than 0.
```

```python
is_sunny = True
if is_sunny:
    print("Wear sunglasses!")  # Runs if is_sunny is True
```

```python
# more than one condition

x = 5
if x < 0:
  print("negative")
elif x == 0:
  print("null")
else: # my_variable > 0
  print("positive")
```

# 7. Loop:
* Definition: A loop is a structure that allows you to repeat a block of code multiple times, based on a condition or a set of items.
* Types:
1. For Loop: Repeats a block of code a fixed number of times.

98a45435-23d3-42c3-b743-70c897c0e5c1_edited_for_loop_syntax.png

```python
for i in range(5):
    print(i)  # Prints 0, 1, 2, 3, 4
```

2. While Loop: Repeats a block of code as long as a condition is true.

while-anatomy.png



```python
while x < 10:
    x += 1  # Increases x by 1 until it reaches 10, equivalent to x = x + 1
```

If the goal is simply to iterate over a list, we can do so directly as follows

```python
grades = [89,62, 99, 85, 97, 65, 54]
```

```python
for grade in grades:
  print(grade)
```

### 7.1 Combining Conditionals and Loops

```python
for grade in grades:
  if grade < 70:
    print("Pass",grade)
```

### 7.2 Loop Control Statements:
* Definition: Special statements that control how a loop operates.
Break: Exits the loop.
* Continue: Skips the rest of the current loop iteration.
Example:



```python
for i in range(5):
    if i == 3:
        break  # Stops the loop when i is 3
    print(i)  # Outputs: 0, 1, 2
```

```python
for i in range(5):
    if i == 3:
        continue  # continue the loop when i is 3 (skip when i is 3)
    print(i)  # Outputs: 0, 1, 2
```

# 8. Functions and Return Statement:
## Function:
* Definition: A function is a block of reusable code that performs a specific task. It takes input (optional),
processes it, and returns an output (optional).

## Return Statement:
* Definition: Used in a function to send the result back to the caller.

Function_anatomy.png

* Example:

```python
def add(a, b):
    return a + b  # Adds two numbers and returns the result

print(add(3, 4))  # Outputs: 7
```

```python
def square(x):
  return x ** 2

print(square(4))  # Outputs: 16
```

```python
def multiply(a, b):
  return a * b

# Functions can be composed.
square(multiply(3, 2))
```

To improve code readability, it is sometimes useful to explicitly name the arguments

```python
square(multiply(a=3, b=2))
```
